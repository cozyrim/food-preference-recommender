```python
import pandas as pd
import random
import numpy as np

# 'mapping.csv' 파일을 읽어옴. 데이터는 cp949로 인코딩되어 있다고 가정
nutrient_df = pd.read_csv('/content/mapping.csv', encoding='cp949')
nutrient_df = nutrient_df.fillna(0)
selected_features = ['식품명','식품분류', '단백질', '지방', '탄수화물', '에너지']
# 반찬 종류 통합
nutrient_df['식품분류'] = nutrient_df['식품분류'].apply(lambda x: '반찬' if '반찬' in x else x)
nutrient_df = nutrient_df[selected_features]
# 확인용 출력
nutrient_df
```

```python
selected_combinations = {}

# 조합 1000개 임의로 선정 (개인용 pc에서 확인을 위한 작업, 한번만 실행)
while len(selected_combinations) < 1000:
  total_calories = 0
  selected_meals = []
# 1200칼로리 이하의 모든 음식 조합 구성
  while total_calories > 1200 or total_calories == 0:
    rice = nutrient_df[nutrient_df['식품분류'] == '밥류'].sample(n=1)
    soup = nutrient_df[nutrient_df['식품분류'] == '국류'].sample(n=1)
    side_dish = nutrient_df[nutrient_df['식품분류'] == '반찬'].sample(n=1)

    total_calories = rice.iloc[0]['에너지'] + soup.iloc[0]['에너지'] + side_dish.iloc[0]['에너지']

    if total_calories <= 1200:
      selected_meals = [rice.iloc[0]['식품명'], soup.iloc[0]['식품명'], side_dish.iloc[0]['식품명']]

  selected_combinations[tuple(selected_meals)] = {'preference': 50, 'total_calories': total_calories}


data = []
for combination, info in selected_combinations.items():
  rice, soup, side_dish = combination
  preference = info['preference']
  total_calories = info['total_calories']

  # '밥', '국', '반찬'을 한 개체로 묶음
  combined_dishes = f"{rice}, {soup}, {side_dish}"

  data.append([combined_dishes, preference, total_calories])

df = pd.DataFrame(data, columns=['식단', '선호도', '칼로리'])
# FoodDataFrame.csv 파일로 저장
df.to_csv("FoodDataFrame.csv", index = False)
```
